```python
#cyclic permutation to check equivalence under the parrtial trace
def cyclic_permutations(lst):
    yield lst
    for k in range(3, len(lst), 3):
        p = lst[k:] + lst[:k]
        if p == lst:
            break
        yield p
```

```python
#cyclic permutation to check equivalence under the parrtial trace
def cyclic_permutations_clock(lst):
    yield lst
    for k in range(3, len(lst), 3):
        p = lst[-k:] + lst[:-k]
        if p == lst:
            break
        yield p
```
